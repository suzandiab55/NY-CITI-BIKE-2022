```python
# Importing necessary libraries
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from datetime import datetime as dt
import seaborn as sns
```

```python
# Reading in csv file for weather
df = pd.read_csv('ny_data.csv', index_col = 0)
```

```python
# Checking data types
df.dtypes
```

## Top 20 Starting Stations Plot

```python
# Assigning each value 1 
df['value'] = 1
```

```python
# Grouping starting stations by their value counts
df_t = df.groupby('start_station_name')['value'].count().reset_index()
```

```python
# Viewing first 5 rows
df_t.head()
```

```python
# Assigning 20 highest value counts to top 20
top20 = df_t.nlargest(20, 'value')
```

```python
# Viewing top 20
top20
```

```python
# Setting theme for plots 
sns.set_theme(style="ticks", palette="Blues_r")
```

```python
# Plotting top 20 starting stations using seaborn
sns.barplot(data = top20, x = 'value', y = 'start_station_name', palette=("Blues_r"))
plt.title("Most Frequent Start Stations")
plt.show()
```

## Dual Axis Plot

```python
df['date'] = pd.to_datetime(df['date'], format = '%Y-%m-%d')
```

```python
# Groupby the count of the trips 

df_group = pd.DataFrame(df.groupby(['date'])['ride_id'].count()).reset_index()
```

```python
# Viewing first 5 rows
df_group.head()
```

```python
# Renaming ride id column, merging data, and printing values 
df_group.rename(columns = {'ride_id':'bike_rides_daily'}, inplace = True)
df = df.merge(df_group, on = "date", how = 'outer', indicator = 'merge_flag')
print(df['merge_flag'].value_counts(dropna = False))
print("Shape of January 1st is", df[df['date'] == '2022-01-01'].shape) # Check 
print("Shape of January 2nd is", df[df['date'] == '2022-01-02'].shape) # Second check 

df_temp = df.set_index('date')

print(df_temp.columns)
```

```python
# Checking columns
df_temp.columns
```

```python
# Viewing first 5 rows
df_temp.head()
```

```python
# Create figure and axis objects with subplots()

fig,ax = plt.subplots(figsize=(10, 5))

# Make a plot of the bike rides
sns.lineplot(data = df_temp['bike_rides_daily'], color = "r")

# Creating the dual axis - initializing a twin object for two different y-axis on the sample plot
ax2 = ax.twinx()

# Make a plot with the temperatures on the y-axis using second axis object
sns.lineplot(data = df_temp['avgTemp'], color="b", ax=ax2)

# Set labels for the axes
ax.set_xlabel("Date", fontsize= 14)
ax.set_ylabel('Bike Rides Daily', color ='red', fontsize= 14)
ax2.set_ylabel('Average Temperature', color='blue', fontsize= 14)

# Add the title
plt.title("Daily Weather and Bike Trips in 2022", fontsize=16)

# Create a combined legend
lines, labels = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc=0)

plt.show()
```

## Box Plot

```python
# Create a box plot for 'avgTemp' grouped by 'member_casual'
plt.figure(figsize=(10, 6))
sns.boxplot(x='member_casual', y='avgTemp', data=df, palette = {'member': 'royalblue', 'casual': 'orange'})
plt.title('Box Plot of Average Temperature by Member Status')
plt.xlabel('Member Status')
plt.ylabel('Average Temperature')
plt.show()
```

The box plot shows that both members and casual users have similar average temperatures. However, casual users experience a wider range of temperatures compared to members. Both groups have similar highest temperatures, but casual users have more variation in their lower temperatures. There are some unusually low temperature readings for both groups. These outliers might be due to specific conditions or errors in data collection. Overall, while both groups have similar temperature patterns, casual users see a bit more variation.

To address these outliers, it is important to inspect and understand their causes:

- Data Validation: Verify if the outliers are a result of data entry or collection errors. Correct any inaccuracies found.
- Contextual Analysis: Determine if the outliers represent legitimate scenarios (e.g., unexpected weather conditions). If so, consider how these scenarios impact overall analysis.
- Outlier Treatment: Decide on a method for handling outliers. This could include removing them from the dataset if they are errors, or applying statistical methods to minimize their impact if they are genuine but extreme values.
- Re-evaluation: After handling the outliers, re-evaluate the box plot and the overall analysis to ensure the results are robust and reliable.

These steps will ensure that the analysis accurately reflects the data and provides meaningful insights."

## Facet Grids

```python
# Parse the date columns
df['started_at'] = pd.to_datetime(df['started_at'])
df['ended_at'] = pd.to_datetime(df['ended_at'])
```

```python
# Extracting additional time-related features
df['hour'] = df['started_at'].dt.hour
df['day_of_week'] = df['started_at'].dt.day_name()
df['month'] = df['started_at'].dt.month_name()
```

```python
# Define the order of months
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
```

```python
# Overall usage trends for Citi Bike in 2022 (by month)
g = sns.FacetGrid(df, col="member_casual", height=6, aspect=1.5)
g.map(sns.histplot, "month", bins=12)
for ax in g.axes.flat:
    ax.set_xticklabels(labels=month_order, rotation=45)
g.set_titles("{col_name} Bike Usage in 2022")
plt.show()
```

Bike usage and ridership was highest in the summer months (June, July, and August) and lowest in February. 

There seems to be a larger disparity in usage between casual and member riders during the colder months. 

This suggests that casual ridership may be more sensitive to weather conditions than member ridership.

FacetGrid plots are useful for visualizing data across multiple subplots, allowing us to compare and analyze different aspects of Citi Bike usage simultaneously.

For instance, they can show how bike usage varies by day of the week and hour of the day, helping to identify peak usage times, or how weather conditions affect bike usage by displaying the number of rides in different temperature bins.

This multi-dimensional approach provides a clearer understanding of patterns and trends.
